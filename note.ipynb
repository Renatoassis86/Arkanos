```python
# Salve como exibir_estrutura_limpa.py na raiz do seu projeto (mesmo nГ­vel de manage.py)

import os

# Pastas/arquivos a ignorar
IGNORAR_PASTAS = {
    "__pycache__", ".git", ".venv", "venv", "env",
    "migrations", "staticfiles", "media", "node_modules"
}
IGNORAR_EXTENSOES = {".pyc", ".db", ".log"}
IGNORAR_ARQUIVOS = {"__init__.py", "db.sqlite3"}

def listar_diretorios(caminho, prefixo=""):
    try:
        arquivos = sorted(os.listdir(caminho))
    except PermissionError:
        return

    for i, nome in enumerate(arquivos):
        caminho_completo = os.path.join(caminho, nome)
        if nome in IGNORAR_ARQUIVOS:
            continue
        if os.path.isdir(caminho_completo):
            if nome in IGNORAR_PASTAS:
                continue
            is_ultimo = i == len(arquivos) - 1
            print(prefixo + ("в””в”Ђв”Ђ " if is_ultimo else "в”њв”Ђв”Ђ ") + nome + "/")
            novo_prefixo = prefixo + ("    " if is_ultimo else "в”‚   ")
            listar_diretorios(caminho_completo, novo_prefixo)
        else:
            _, ext = os.path.splitext(nome)
            if ext in IGNORAR_EXTENSOES:
                continue
            print(prefixo + ("в””в”Ђв”Ђ " if i == len(arquivos) - 1 else "в”њв”Ђв”Ђ ") + nome)

if __name__ == "__main__":
    print("рџ“Ѓ Estrutura limpa do seu projeto:")
    listar_diretorios(".")
```
